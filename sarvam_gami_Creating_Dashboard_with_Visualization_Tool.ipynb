1. What is Power BI and how does it differ from Excel?
Power BI is a business analytics tool by Microsoft used to visualize data, create interactive dashboards, and share reports.
It differs from Excel in that Power BI is designed for scalable data models, automated refresh, interactive reports, and real-time data, while Excel is primarily used for manual data analysis, spreadsheets, and pivot tables.

2. Explain the concept of data modeling in Power BI.
Data modeling in Power BI involves creating relationships between tables to allow efficient querying and accurate reporting. It includes defining keys, establishing one-to-many or many-to-one relationships, creating calculated columns, measures, hierarchies, and optimizing models for performance and usability.

3. What are the different types of connections available in Power BI?
Power BI supports these connection types:

Import: Data is loaded into Power BI’s memory (fastest performance).

DirectQuery: Queries data live without importing.

Live Connection: Used with SQL Server Analysis Services or Power BI datasets.

Composite Model: Combines Import and DirectQuery in one model.

4. How do you handle data transformation in Power BI?
Data transformation is done in the Power Query Editor. You can:

Remove or filter rows

Change data types

Merge or append tables

Pivot/unpivot columns

Split columns
These transformations clean and shape the data before loading it into the model.

5. What is DAX (Data Analysis Expressions) and why is it important in Power BI?
DAX is a formula language used to perform calculations in Power BI. It’s essential for:

Creating measures for aggregated data

Building calculated columns and tables

Applying business logic and complex filters
DAX enables dynamic and powerful data analysis.

6. Can you explain the difference between calculated columns and measures in Power BI?
Calculated columns are computed row-by-row and stored in the table; used for filtering and slicing.

Measures are calculated on-the-fly based on the filter context; used for aggregations like sum, average, etc.
Measures are more efficient and better for performance compared to calculated columns.

7. How do you handle relationships between tables in Power BI?
Relationships are managed in Model View, where you can:

Define one-to-many or many-to-one relationships

Set cross-filter direction

Ensure data integrity using primary and foreign keys
This enables accurate data aggregation across multiple tables.

8. What is the purpose of a Power BI Gateway?
A Power BI Gateway bridges cloud services and on-premises data sources.

Personal Gateway: For individual users; supports refresh only.

Enterprise Gateway: Supports multiple users, live queries, and scheduled refreshes.
It’s essential for keeping cloud reports synced with local data sources.

9. How can you schedule data refresh in Power BI Service?
In Power BI Service:

Go to your dataset settings.

Enable Scheduled Refresh.

Set the refresh frequency and time.

Enter credentials and ensure a gateway is configured if needed.
This keeps reports updated automatically.

10. Explain the concept of row-level security in Power BI.
Row-Level Security (RLS) limits data visibility based on the user.

You define roles and DAX filters in Power BI Desktop.

Assign users to roles in Power BI Service.
RLS ensures users only see data relevant to them, enhancing security.

11. What is the Power BI Desktop and how does it differ from Power BI Service?
Power BI Desktop: A Windows application used to design, model, and transform data.

Power BI Service: A cloud-based platform used to publish, share, collaborate, and refresh reports.
Desktop is for development, while Service is for sharing and consumption.

12. Explain the concept of Direct Query in Power BI.
DirectQuery allows Power BI to connect directly to the data source without importing the data.

Data is queried in real time.

Supports large datasets.

Slower than import mode and limited in DAX functions.
Used when data changes frequently or needs to be up-to-date.

13. What are Power BI templates and how are they useful?
Power BI templates (.PBIT files) contain the report layout, data model, and queries—without actual data.
Useful for:

Sharing report designs

Creating report frameworks

Reusing standardized layouts across environments

14. How do you handle incremental data refresh in Power BI?
Steps for incremental refresh:

Create RangeStart and RangeEnd parameters.

Filter your data using those parameters.

Enable Incremental Refresh in table settings.
This optimizes performance by only refreshing new or changed data.

15. What is the role of Power Query in Power BI?
Power Query is the ETL (Extract, Transform, Load) tool in Power BI.
It’s used to:

Connect to various data sources

Clean, transform, and shape data

Apply logic before data is loaded to the model
It uses the M language behind the scenes.

16. Explain the difference between calculated columns and calculated tables in Power BI.
Calculated Columns: Add new columns to existing tables; evaluated row-by-row.

Calculated Tables: Create entirely new tables using DAX; useful for summarized or filtered data sets.
Both are stored in the model but serve different purposes.

17. How do you create custom visuals in Power BI?
You can create custom visuals by:

Using Power BI Developer Tools and TypeScript

Creating visuals via the Power BI Visuals SDK

Importing visuals from the AppSource marketplace
These visuals are useful when standard charts don’t meet requirements.

18. What are the best practices for optimizing performance in Power BI?
Use Import mode where possible.

Reduce data size (remove unused columns/rows).

Use measures over calculated columns.

Optimize DAX queries.

Limit visuals per report page.

Use aggregation tables and star schema design.

Enable incremental refresh for large datasets.

19. How can you integrate Power BI with other Microsoft products like Azure and Office 365?
Azure: Use services like Azure SQL, Azure Data Lake, Synapse Analytics.

Office 365: Embed reports in Teams, SharePoint, or integrate with Excel Online.

Use Power Automate to trigger workflows based on Power BI alerts.

Use Microsoft Fabric for end-to-end data integration.

20. Explain the concept of aggregations in Power BI.
Aggregations are pre-summarized tables created to optimize performance:

Reduce query time for large datasets

Can be automatic or manual

Linked to detailed data via relationships
Used in composite models to improve report performance without sacrificing granularity.

21. How do you handle error handling and data quality in Power BI?
In Power BI:

Use Power Query to detect and fix errors (e.g., nulls, data types).

Apply filters, conditional columns, or validations.

Use error-catching DAX functions like IFERROR().

Profile columns to check for anomalies.

Maintain clean, consistent source data.

22. What is the purpose of Power BI Embedded and when would you use it?
Power BI Embedded allows developers to embed Power BI reports and dashboards into custom applications or websites.
Use cases:

SaaS applications

Internal portals

White-label analytics solutions
It enables analytics-as-a-service without requiring users to access Power BI directly.
